# **ADMET Filtering**

ADMET filtering is a crucial step in the drug discovery process that involves evaluating the Absorption, Distribution, Metabolism, Excretion, and Toxicity (ADMET) properties of compounds. This helps in identifying potential drug candidates that are not only effective but also safe for human use.

```python
!pip install rdkit-pypi
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
import matplotlib.pyplot as plt
```

```python
def Filter(DataFrame):
  NewDataFrame = pd.DataFrame()
  NewDataFrame = DataFrame[(DataFrame['logP']>=1) & (DataFrame['logP']<=5)
                         & (DataFrame['Solubility_AqSolDB']>-6) & (DataFrame['Solubility_AqSolDB']<=-4)
                         & (DataFrame['BBB_Martins']>0.5)
                         & (DataFrame['Lipinski']==4)
                         & (DataFrame['ClinTox']<0.5)
                         & (DataFrame['PAINS'] == '[]')
                         & (DataFrame['sa_score']<5)]
  return NewDataFrame
```

# **PubChem**

## *(4-hydroxynaphthalen-1-yl)sulfonamide*

```python
url = 'https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/ADMET_analysis_p3/PubChem_sulfonamide_ADMET_p3.csv'
df_PubChem_sulfonamide_ADMET_p3 = pd.read_csv(url)
df_PubChem_sulfonamide_ADMET_p3.shape[0]
```

```python
df_PubChem_sulfonamide_ADMET_temp = Filter(df_PubChem_sulfonamide_ADMET_p3)
df_PubChem_sulfonamide_ADMET_filtered = pd.DataFrame()
df_PubChem_sulfonamide_ADMET_filtered['smiles'] = df_PubChem_sulfonamide_ADMET_temp['smiles']
df_PubChem_sulfonamide_ADMET_filtered.shape[0]
```

```python
df_PubChem_sulfonamide_ADMET_filtered.to_csv('PubChem_sulfonamide_ADMET_filtered.csv', index=False, header=False)
```
