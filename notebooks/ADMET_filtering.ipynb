```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def Filter(DataFrame):
  NewDataFrame = pd.DataFrame()
  NewDataFrame = DataFrame[(DataFrame['logP']>=1) & (DataFrame['logP']<=5)
                         & (DataFrame['Solubility_AqSolDB']>=-6) & (DataFrame['Solubility_AqSolDB']<=-4)
                         & (DataFrame['BBB_Martins']>0.5)
                         & (DataFrame['Lipinski']==4)
                         & (DataFrame['ClinTox']<0.5)]
  return NewDataFrame
```

```python
DataFrame = pd.read_csv('https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/ADMET_analysis/PubChem_stilbene_ADMET.csv')
DataFrame
```

```python
DF = Filter(DataFrame)
DF2 = pd.DataFrame()
DF2['smiles'] = DF['smiles']
DF2.to_csv('out.csv', index=False)
```
