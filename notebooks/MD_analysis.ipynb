```python
PL_RMSD_2 = pd.read_csv('PL_RMSD.dat', sep='\s+')
PL_RMSD_3 = pd.read_csv('PL_RMSD.dat.1', sep='\s+')
PL_RMSD_4 = pd.read_csv('PL_RMSD.dat.2', sep='\s+')
```

```python
sns.set_style("white")
sns.set_style("ticks")
plt.figure(figsize=(8, 4), dpi=200)

sns.lineplot(data=PL_RMSD_2, x=PL_RMSD_2['#_frame#']/10, y='Prot_CA', label='UNPD82299', color='c')
sns.lineplot(data=PL_RMSD_3, x=PL_RMSD_3['#_frame#']/10, y='Prot_CA', label='UNPD96223', color='m')
sns.lineplot(data=PL_RMSD_4, x=PL_RMSD_4['#_frame#']/10, y='Prot_CA', label='UNPD228592', color='y')

plt.title('RMSD de Proteina - Cα',fontweight='bold')
plt.xlabel('Tiempo (ns)', color='gray')
plt.ylabel('RMSD ($\it{Å}$)', color='gray')
plt.legend(loc="lower right")
#plt.savefig('rmsd_MeDFO.png', format='png', dpi=350)
plt.show()
```

```python
sns.set_style("white")
sns.set_style("ticks")
plt.figure(figsize=(8, 4), dpi=200)

sns.lineplot(data=PL_RMSD_2, x=PL_RMSD_2['#_frame#']/10, y='Lig_wrt_Protein', label='UNPD82299', color='c')
sns.lineplot(data=PL_RMSD_3, x=PL_RMSD_3['#_frame#']/10, y='Lig_wrt_Protein', label='UNPD96223', color='m')
sns.lineplot(data=PL_RMSD_4, x=PL_RMSD_4['#_frame#']/10, y='Lig_wrt_Protein', label='UNPD228592', color='y')

plt.title('RMSD de Ligando',fontweight='bold')
plt.xlabel('Tiempo (ns)', color='gray')
plt.ylabel('RMSD ($\it{Å}$)', color='gray')
plt.legend(loc="lower right")
#plt.savefig('rmsd_MeDFO.png', format='png', dpi=350)
plt.show()
```
