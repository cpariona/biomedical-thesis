# **PubChem**

## *(4-hydroxynaphthalen-1-yl)sulfonamide*

```python
!wget https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/ADMET_filtered/PubChem_sulfonamide_ADMET_filtered.csv
os.rename('PubChem_sulfonamide_ADMET_filtered.csv', 'PubChem_sulfonamide_ADMET_filtered.smi')
```

```python
!obabel -i smi PubChem_sulfonamide_ADMET_filtered.smi -o sdf --gen3D > _.sdf
!obabel -i sdf _.sdf -o mol2 -O PubChem_sulfonamide_.mol2 -h -m
!obabel *.mol2 -O opt.xyz --minimize --sd --steps 20000 --ff MMFF94 -p 7.4 --log -m
!obabel -i xyz *.xyz -o pdbqt -O lig_.pdbqt -m
!rm *.mol2 *.xyz *.sdf
!mkdir PubChem_sulfonamide
!mv *.pdbqt PubChem_sulfonamide
!zip -r PubChem_sulfonamide.zip PubChem_sulfonamide
!rm -r PubChem_sulfonamide
```
