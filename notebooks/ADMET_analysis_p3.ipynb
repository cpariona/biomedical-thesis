```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import RDConfig
import os
import sys
sys.path.append(os.path.join(RDConfig.RDContribDir, 'SA_Score'))
import sascorer
```

```python
def sa_score(smiles):

  mol = Chem.MolFromSmiles(smiles)
  s = sascorer.calculateScore(mol)

  return s
```

# **PubChem**

## *(4-hydroxynaphthalen-1-yl)sulfonamide*

```python
url = 'https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/ADMET_analysis_p2/PubChem_sulfonamide_ADMET_p2.csv'
df_PubChem_sulfonamide_ADMET_p2 = pd.read_csv(url)
```

```python
df_PubChem_sulfonamide_ADMET_p2['sa_score'] = [sa_score(df_PubChem_sulfonamide_ADMET_p2['smiles'][i]) for i in range(len(df_PubChem_sulfonamide_ADMET_p2['smiles']))]
```

```python
df_PubChem_sulfonamide_ADMET_p2.to_csv('PubChem_sulfonamide_ADMET_p3.csv', index=False)
```
