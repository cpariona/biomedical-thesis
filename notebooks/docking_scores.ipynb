```python
# Import libraries
import pandas as pd
```

```python
# Import databases
PubChem_sulfonamide_df = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/refs/heads/main/data/ADMET_filtered/PubChem_sulfonamide_ADMET_filtered.csv", header=None)
PubChem_indolinone_df = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/refs/heads/main/data/ADMET_filtered/PubChem_indolinone_ADMET_filtered.csv", header=None)
PubChem_flavanol_df = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/refs/heads/main/data/ADMET_filtered/PubChem_flavanol_ADMET_filtered.csv", header=None)
PubChem_stilbene_df = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/refs/heads/main/data/ADMET_filtered/PubChem_stilbene_ADMET_filtered.csv", header=None)
DrugBank_indolinone_df = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/refs/heads/main/data/ADMET_filtered/DrugBank_indolinone_ADMET_filtered.csv", header=None)
DrugBank_stilbene_df = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/refs/heads/main/data/ADMET_filtered/DrugBank_stilbene_ADMET_filtered.csv", header=None)
UNPD_indolinone_df = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/refs/heads/main/data/ADMET_filtered/UNPD_indolinone_ADMET_filtered.csv", header=None)
UNPD_flavanol_df = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/refs/heads/main/data/ADMET_filtered/UNPD_flavanol_ADMET_filtered.csv", header=None)
UNPD_stilbene_df = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/refs/heads/main/data/ADMET_filtered/UNPD_stilbene_ADMET_filtered.csv", header=None)
```

```python
# Import data docking results
PubChem_sulfonamide_vina = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/PubChem_sulfonamide/resumen_VS.txt", sep="    ", header=None)
PubChem_indolinone_vina = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/PubChem_indolinone/resumen_VS.txt", sep="    ", header=None)
PubChem_flavanol_vina = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/PubChem_flavanol/resumen_VS.txt", sep="    ", header=None)
PubChem_stilbene_vina = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/PubChem_stilbene/resumen_VS.txt", sep="    ", header=None)
DrugBank_indolinone_vina = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/DrugBank_indolinone/resumen_VS.txt", sep="    ", header=None)
DrugBank_stilbene_vina = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/DrugBank_stilbene/resumen_VS.txt", sep="    ", header=None)
UNPD_indolinone_vina = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/UNPD_indolinone/resumen_VS.txt", sep="    ", header=None)
UNPD_flavanol_vina = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/UNPD_flavanol/resumen_VS.txt", sep="    ", header=None)
UNPD_stilbene_vina = pd.read_csv("https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/UNPD_stilbene/resumen_VS.txt", sep="    ", header=None)
```

```python
# Define functions
def extract_docking_mol_idx(df_vina):
  df_vina[['./lig','opt','DB','Mol','idx.pdbqt.log']] = df_vina[0].str.split('_', expand=True)
  df_vina[['idx','pdbqt','log']] = df_vina['idx.pdbqt.log'].str.split('.', expand=True)
  df_vina.drop(columns=['./lig','opt','DB','Mol','idx.pdbqt.log','pdbqt','log'], inplace=True)
  df_vina['idx'] = df_vina['idx'].astype(int) - 1
  df_vina.rename({1: 'Delta_Gibbs'}, axis=1, inplace=True)
  return df_vina

def smiles_docking_merge(df_smiles, df_docking):
  df_smiles['idx'] = df_smiles.index
  df_merged = pd.merge(df_smiles, df_docking[['Delta_Gibbs', 'idx']], on='idx', how='inner')
  df_merged_result = df_merged[[0, 'Delta_Gibbs']].copy()
  df_merged_result = df_merged_result.rename(columns={0: 'smiles'})
  return df_merged_result
```

```python
# Extract docking results index
PubChem_sulfonamide_docking = extract_docking_mol_idx(PubChem_sulfonamide_vina)
PubChem_indolinone_docking = extract_docking_mol_idx(PubChem_indolinone_vina)
PubChem_flavanol_docking = extract_docking_mol_idx(PubChem_flavanol_vina)
PubChem_stilbene_docking = extract_docking_mol_idx(PubChem_stilbene_vina)
DrugBank_indolinone_docking = extract_docking_mol_idx(DrugBank_indolinone_vina)
DrugBank_stilbene_docking = extract_docking_mol_idx(DrugBank_stilbene_vina)
UNPD_indolinone_docking = extract_docking_mol_idx(UNPD_indolinone_vina)
UNPD_flavanol_docking = extract_docking_mol_idx(UNPD_flavanol_vina)
UNPD_stilbene_docking = extract_docking_mol_idx(UNPD_stilbene_vina)
```

```python
# Merge smiles with delta Gibbs
PubChem_sulfonamide_scores = smiles_docking_merge(PubChem_sulfonamide_df,PubChem_sulfonamide_docking)
PubChem_indolinone_scores = smiles_docking_merge(PubChem_indolinone_df,PubChem_indolinone_docking)
PubChem_flavanol_scores = smiles_docking_merge(PubChem_flavanol_df,PubChem_flavanol_docking)
PubChem_stilbene_scores = smiles_docking_merge(PubChem_stilbene_df,PubChem_stilbene_docking)
DrugBank_indolinone_scores = smiles_docking_merge(DrugBank_indolinone_df,DrugBank_indolinone_docking)
DrugBank_stilbene_scores = smiles_docking_merge(DrugBank_stilbene_df,DrugBank_stilbene_docking)
UNPD_indolinone_scores = smiles_docking_merge(UNPD_indolinone_df,UNPD_indolinone_docking)
UNPD_flavanol_scores = smiles_docking_merge(UNPD_flavanol_df,UNPD_flavanol_docking)
UNPD_stilbene_scores = smiles_docking_merge(UNPD_stilbene_df,UNPD_stilbene_docking)
```

```python
scores_df = pd.concat([PubChem_sulfonamide_scores,
                       PubChem_indolinone_scores,
                       PubChem_flavanol_scores,
                       PubChem_stilbene_scores,
                       DrugBank_indolinone_scores,
                       DrugBank_stilbene_scores,
                       UNPD_indolinone_scores,
                       UNPD_flavanol_scores,
                       UNPD_stilbene_scores], ignore_index=True)
```

```python
scores_df.to_csv('docking_scores_data.csv', index=False)
```
