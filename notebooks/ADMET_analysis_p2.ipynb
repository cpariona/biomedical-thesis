```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
import pandas as pd
```

```python
!wget https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/ADMET_analysis_p2/pains.txt
inf = open('pains.txt', "r")
sub_strct = [ line.rstrip().split(" ") for line in inf ]

smarts = [ line[0] for line in sub_strct]
desc = [ line[1] for line in sub_strct]
dic = dict(zip(smarts, desc))

def pains(smiles):

    mol = Chem.MolFromSmiles(smiles)

    for k,v in dic.items():
        subs = Chem.MolFromSmarts( k )
        if subs != None:
            if mol.HasSubstructMatch( subs ):
                mol.SetProp(v,k)

    props = [prop for prop in mol.GetPropNames()]

    return props
```

# **PubChem**

## *(4-hydroxynaphthalen-1-yl)sulfonamide*

```python
url = 'https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/ADMET_analysis_p1/PubChem_sulfonamide_ADMET_p1.csv'
df_PubChem_sulfonamide_ADMET_p1 = pd.read_csv(url)
```

```python
df_PubChem_sulfonamide_ADMET_p1['PAINS'] = [pains(df_PubChem_sulfonamide_ADMET_p1['smiles'][i]) for i in range(len(df_PubChem_sulfonamide_ADMET_p1['smiles']))]
```

```python
df_PubChem_sulfonamide_ADMET_p1.to_csv('PubChem_sulfonamide_ADMET_p2.csv', index=False)
```
