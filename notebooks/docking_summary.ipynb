```python
# Import libraries
import pandas as pd
from rdkit import Chem
from rdkit.Chem import PandasTools
from rdkit.Chem import Draw
from rdkit.Chem.Draw import MolsToGridImage
from rdkit.Chem.Draw import SimilarityMaps
```

```python
# Import data files
url1 = "https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/PubChem_sulfonamide/resumen_VS.txt"
url2 = "https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/PubChem_indolinone/resumen_VS.txt"
url3 = "https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/PubChem_flavanol/resumen_VS.txt"
url4 = "https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/PubChem_stilbene/resumen_VS.txt"
url5 = "https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/DrugBank_indolinone/resumen_VS.txt"
url6 = "https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/DrugBank_stilbene/resumen_VS.txt"
url7 = "https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/UNPD_indolinone/resumen_VS.txt"
url8 = "https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/UNPD_flavanol/resumen_VS.txt"
url9 = "https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/docking_results/UNPD_stilbene/resumen_VS.txt"
PubChem_sulfonamide = pd.read_csv(url1, sep="    ", header=None)
PubChem_indolinone = pd.read_csv(url2, sep="    ", header=None)
PubChem_flavanol = pd.read_csv(url3, sep="    ", header=None)
PubChem_stilbene = pd.read_csv(url4, sep="    ", header=None)
DrugBank_indolinone = pd.read_csv(url5, sep="    ", header=None)
DrugBank_stilbene = pd.read_csv(url6, sep="    ", header=None)
UNPD_indolinone = pd.read_csv(url7, sep="    ", header=None)
UNPD_flavanol = pd.read_csv(url8, sep="    ", header=None)
UNPD_stilbene = pd.read_csv(url9, sep="    ", header=None)
```

```python
# Sort by Delta G
PubChem_sulfonamide.sort_values(by=1, ascending=True, inplace=True)
PubChem_indolinone.sort_values(by=1, ascending=True, inplace=True)
PubChem_flavanol.sort_values(by=1, ascending=True, inplace=True)
PubChem_stilbene.sort_values(by=1, ascending=True, inplace=True)
DrugBank_indolinone.sort_values(by=1, ascending=True, inplace=True)
DrugBank_stilbene.sort_values(by=1, ascending=True, inplace=True)
UNPD_indolinone.sort_values(by=1, ascending=True, inplace=True)
UNPD_flavanol.sort_values(by=1, ascending=True, inplace=True)
UNPD_stilbene.sort_values(by=1, ascending=True, inplace=True)
```

```python
# Select top 5
PubChem_sulfonamide_top = PubChem_sulfonamide.head(5)
PubChem_indolinone_top = PubChem_indolinone.head(5)
PubChem_flavanol_top = PubChem_flavanol.head(5)
PubChem_stilbene_top = PubChem_stilbene.head(5)
DrugBank_indolinone_top = DrugBank_indolinone.head(5)
DrugBank_stilbene_top = DrugBank_stilbene.head(5)
UNPD_indolinone_top = UNPD_indolinone.head(5)
UNPD_flavanol_top = UNPD_flavanol.head(5)
UNPD_stilbene_top = UNPD_stilbene.head(5)
```

```python
results = pd.concat([PubChem_sulfonamide_top,
                     PubChem_indolinone_top,
                     PubChem_flavanol_top,
                     PubChem_stilbene_top,
                     DrugBank_indolinone_top,
                     DrugBank_stilbene_top,
                     UNPD_indolinone_top,
                     UNPD_flavanol_top,
                     UNPD_stilbene_top], ignore_index=True)
```
