# **PeruNPDB**

```python
url = 'https://raw.githubusercontent.com/cpariona/biomedical-thesis/main/data/raw/PeruNPDB.csv'
df_PeruNPDB = pd.read_csv(url, names=['SMILES'])
```

```python
mol_list = [Chem.MolFromSmiles(x) for x in df_PeruNPDB['SMILES']]
query = Chem.MolFromSmarts("CC")
match_list = []
for mol in mol_list:
  try:
    match_list.append(mol.GetSubstructMatch(query))
  except:
    print(mol)

# MolsToGridImage(mols=mol_list, molsPerRow=4,highlightAtomLists=match_list)
```
