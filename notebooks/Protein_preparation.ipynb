```python
#@title Visualizacion de ALA10.pdb

import py3Dmol

view = py3Dmol.view()
view.addModel(open('1xq8_ph.gro', 'r').read(),'gro')
view.setBackgroundColor('black')
view.setStyle({'chain':'A'}, {'stick': {'colorscheme':'yellowCarbon'}})
view.zoomTo()
view.show()
```
